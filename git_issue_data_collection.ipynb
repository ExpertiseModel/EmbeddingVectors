## Collect Issue dataset

```python
class GitIssueScrapper:

    def __init__(self):
        auth_git = #put your github authentication key here
        self.g = Github(auth_git)

    def issue_info_extracter(self, users):

        #users_iter = iter(users)

        for u in users:
            try:
                user = self.g.get_user(u)
                try:
                    name = user.name

                except:
                    name = "none"
                    
                try:
                    u_id = user.login

                except:
                    u_id = "none"


                try:
                    reps = user.get_repos()
                except:
                    reps = "none"

                all_issue_repo={}
                counter=0
                for repository in reps:
                    full_issue={}
                    cnt=0
                    try:
                        repo_parent = repository.parent
                    except:
                        repo_parent = "None"

         

                    try:
                        if repo_parent:
                            mention_issues=repository.parent.get_issues(mentioned=user,state='all')
                            create_issues=repository.parent.get_issues(creator=user,state='all')
                            assigne_issues=repository.parent.get_issues(assignee=user,state='all')

                        else:
                            mention_issues=repository.get_issues(mentioned=user,state='all')
                            create_issues=repository.get_issues(creator=user,state='all')
                            assigne_issues=repository.get_issues(assignee=user,state='all')
                    except:
                        issues = []

                    for issue in mention_issues:
                        full_issue[cnt] = {'title':issue.title,'body':issue.body}
                        cnt += 1
    
                    for issue in create_issues:
                        full_issue[cnt] = {'title':issue.title,'body':issue.body}
                        cnt += 1
    
                    for issue in assigne_issues:
                        full_issue[cnt] = {'title':issue.title,'body':issue.body}
                        cnt += 1

                    all_issue_repo[counter] = full_issue
                    counter += 1

                    
                        
    


                
                data = {}

                data['name'] = name
                data['u_id'] = u_id
                data['issues'] = all_issue_repo

                    
        
                pickle.dump(data, open(('../' + u_id + '.pickle'), 'wb'))
            

                print(u_id, "-------------------> done")

            except RateLimitExceededException:
                now = datetime.now()
                current_time = now.strftime("%H:%M:%S")
                search_rate_limit = self.g.get_rate_limit().search
                print('current time:{}'.format(current_time))
                #time.sleep(sleep_time)
                time.sleep(3600)
                continue
            except UnknownObjectException:
                continue
                
    def get_issue_details(self):
        # users = self.git.search_users('', location=self.location, language=self.language)
        #users = self.git.search_users('', location='montreal')
        self.issue_info_extracter(user_roles)  
```
